```python
import altair as alt
import polars as pl

# Compute normalized counts (likelihood)
hist = (
    subset.lazy()
    .sort("AgeDays")
    .group_by_dynamic("AgeDays", every="7i")  # or use pl.col("AgeDays").floor_div(1)
    .agg(pl.count().alias("count"))
    .collect()
)

total = hist["count"].sum()
hist = hist.with_columns((pl.col("count") / total).alias("likelihood"))

alt.Chart(hist).mark_bar().encode(
    x=alt.X("AgeDays:Q", bin=alt.Bin(step=1), scale=alt.Scale(domain=[0, 150])),
    y=alt.Y("likelihood:Q", title="Likelihood"),
).properties(
    title=f"AgeDays likelihood for OutcomeType = {outcome}",
    height=50,
    width=1000
).show()
```

```python
import altair as alt
import polars as pl

for outcome in (
    df.filter(pl.col("OutcomeType").is_not_null())
      .select(pl.col("OutcomeType").unique())
      .to_series()
      .to_list()
):
    subset = df.filter(pl.col("OutcomeType") == outcome, pl.col("AgeDays").is_not_null())

    # Compute normalized counts (likelihood)
    hist = (
        subset.lazy()
        .sort("AgeDays")
        .group_by_dynamic("AgeDays", every="7i")  # or use pl.col("AgeDays").floor_div(1)
        .agg(pl.count().alias("count"))
        .collect()
    )

    total = hist["count"].sum()
    hist = hist.with_columns((pl.col("count") / total).alias("likelihood"))

    alt.Chart(hist).mark_bar().encode(
        x=alt.X("AgeDays:Q", bin=alt.Bin(step=1), scale=alt.Scale(domain=[0, 150])),
        y=alt.Y("likelihood:Q", title="Likelihood"),
    ).properties(
        title=f"AgeDays likelihood for OutcomeType = {outcome}",
        height=50,
        width=1000
    ).show()
```
