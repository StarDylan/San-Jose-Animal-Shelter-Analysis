```python
import polars as pl
import altair as alt
alt.data_transformers.enable("vegafusion")
```

```python
df = pl.read_parquet("data/clean/cleaned_data.parquet")

df.plot.bar(
    alt.X("IntakeMedicalIssueIndex", bin=alt.Bin(step=1)),
    y='count()',
)
```

```python
from turtle import title


for index_val in range(5):
    subset = df.filter(pl.col("IntakeMedicalIssueIndex") == index_val)
    subset.plot.bar(
        alt.X("TimeInShelterDays", bin=alt.Bin(step=1), scale=alt.Scale(domain=[0, 150])),
        y='count()',
    ).properties(title=f"Time in Shelter for IntakeMedicalIssueIndex = {index_val}", height=50, width = 1000).show()
```

```python
from turtle import title


for index_val in df.filter(pl.col("OutcomeType").is_not_null()).select(pl.col("OutcomeType").unique()).to_series().to_list():
    subset = df.filter(pl.col("OutcomeType") == index_val)
    subset.plot.bar(
        alt.X("AgeDays", bin=alt.Bin(step=1), scale=alt.Scale(domain=[0, 150])),
        y='count()',
    ).properties(title=f"AgeDays for OutcomeType = {index_val}", height=50, width = 1000).show()
```
